```python
import pandas as pd
import os
import re
```

```python
directory = '../data/'
all_files = [os.path.join(directory, f) for f in os.listdir(directory) if f.endswith('.csv')]
```

```python
print(all_files)
```

```python
def extract_number(file_path):
    match = re.search(r'(\d+)', file_path)
    return int(match.group(1)) if match else -1
sorted_file_list = sorted(all_files, key=extract_number)
```

```python
store_df = pd.read_csv('../store.csv')
```

```python
print(store_df)
```

```python
first_column_name = store_df.columns[0]

# ì²« ë²ˆì§¸ ì»¬ëŸ¼ì„ ì¸ë±ìŠ¤ì™€ ê°’ìœ¼ë¡œ ë”•ì…”ë„ˆë¦¬ ë³€í™˜
store_dict = {index: value for index, value in enumerate(store_df[first_column_name])}

# ê²°ê³¼ ì¶œë ¥
print(store_dict)
```

```python
store_dict[0]
```

```python
# ì›í•˜ëŠ” ì»¬ëŸ¼ ì´ë¦„
target_column = 'text'

# ë¹ˆ ë°ì´í„°í”„ë ˆì„ ë¦¬ìŠ¤íŠ¸ ìƒì„±
dataframes = []

# íŒŒì¼ë“¤ì„ ì½ì–´ì„œ ìƒˆë¡œìš´ ë°ì´í„°í”„ë ˆì„ êµ¬ì„±
for file in sorted_file_list:
    # íŒŒì¼ ì¸ë±ìŠ¤ë¥¼ ì¶”ì¶œ (ì˜ˆ: 'data/0.csv'ì—ì„œ 0ì„ ì¶”ì¶œ)
    file_index = int(file.split('/')[-1].split('.')[0])
    # í•´ë‹¹ íŒŒì¼ ì½ê¸°
    df = pd.read_csv(file, usecols=[target_column])
    
    # ìƒˆë¡œìš´ ë°ì´í„°í”„ë ˆì„ ìƒì„±
    new_df = pd.DataFrame({
        'store_name': store_dict[file_index],
        target_column: df[target_column]
    })
    
    # ë°ì´í„°í”„ë ˆì„ ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€
    dataframes.append(new_df)
```

```python
combined_df = pd.concat(dataframes, ignore_index=True)
```

```python
print(combined_df)
```

```python
combined_df.to_csv('combined_data.csv', index=False)
```
